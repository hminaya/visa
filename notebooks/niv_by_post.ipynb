```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fl = 'data/consolidated/niv_by_post.csv'
```

```python
niv = pd.read_csv(fl, delimiter="|", dtype={'Quantity': int})
```

```python
niv
```

```python
niv.loc[:,['Post']]
```

```python
niv.groupby('Post').head(3).sort_values(['Quantity'], ascending=False)
```

```python
niv.groupby('Post').sum().sort_values(['Quantity'], ascending=False)
```

```python
niv.groupby(['Post', 'Year']).sum().sort_values(['Post','Quantity'], ascending=False)
```

```python
niv.groupby('Post').sum().sort_values(['Quantity'], ascending=False).head(10)['Quantity'].plot()
```
